```python
import pandas as pd
import json
import nltk
from nltk.tokenize import word_tokenize
import string
# Load the CSV file
file_path = 'cleaned_data.csv'
data = pd.read_csv(file_path)

data
```

```python
correct_formatted_pairs = []

# Iterate over each row in the DataFrame to correctly format the question and answer pairs
for index, row in data.iterrows():
    corrected_formatted_question = f"</start> {row['question']} </s> {row['answer']} </end> "
    # Append the correctly formatted pair
    correct_formatted_pairs.append(corrected_formatted_question)

# Combine all pairs into a single graph representation with the corrected format
correct_graph_representation = "".join(correct_formatted_pairs)

# Encode the corrected graph representation to ASCII code
# correct_ascii_encoded_graph = [ord(char) for char in correct_graph_representation]

# Create a new JSON file with the corrected ASCII encoded graph
correct_encoded_graph_path = 'clean_data/gathered_graph.json'
with open(correct_encoded_graph_path, 'w') as file:
    json.dump(correct_graph_representation, file)

correct_encoded_graph_path
```
